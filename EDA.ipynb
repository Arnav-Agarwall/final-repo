# Import required libbraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# reading the file

```python
data= pd.read_csv('adult.brl.csv')
```

```python
data
```

# basic information of dataset

# dataset size

```python
data.shape
```

# identifying null data

```python
data=data.replace('?',np.nan)
```

```python
data
```

```python
data.isnull().sum()
```

#### since null data is very less
# dropping null data

```python
data = data.dropna()
```

```python
data
```

# Finding duplicate data

```python
dupl=data.duplicated().sum()
print(f"{dupl}")
```

# Dropping duplicate data

```python
data = data.drop_duplicates()
```

```python
income = data['income'].value_counts()
plt.figure(figsize=(12, 12))
plt.pie(income,labels=income.index, autopct='%1.1f%%', startangle=90 )
plt.title('Income Distribution')
plt.show()

plt.figure(figsize=(12, 6))
sns.countplot(x='education', data=data)
plt.title('Education Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='income', y='age', data=data,width=0.5)
plt.title('Age Distribution by Income')
plt.show()

plt.figure(figsize=(12, 6))
sns.countplot(x='education', hue='income', data=data, width=0.5)
plt.title('Income by Education Level')
plt.xticks(rotation=45)
plt.show()
```

```python
num_cols = data.select_dtypes(include=['int64']).columns

data[num_cols].hist(figsize=(12, 10), bins=15)
plt.tight_layout()
plt.show()
```

```python
correlation_matrix = data[num_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
category = data.select_dtypes(include=['object']).columns

for col in category:
    print(f"Value counts for {col}:\n", data[col].value_counts())
    print("\n")
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='marital.status', hue='income', data=data)
plt.title('Income Distribution by Marital Status')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='age', hue='income', bins=25, multiple='stack')
plt.title('Age Distribution by Income Level')
plt.xlabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='income', y='hours.per.week', data=data, width=0.5)
plt.title('Hours Worked per Week by Income Level')
plt.show()


plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='hours.per.week', hue='income', bins=30, multiple='stack')
plt.title('Hours Worked per Week by Income Level')
plt.xlabel('Hours per Week')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='education', hue='income', data=data)
plt.title('Income Level by Education')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='sex', hue='income', data=data)
plt.title('Income Distribution by Gender')
plt.show()
```

```python
g = sns.FacetGrid(data, col='income', height=5)
g.map(sns.histplot, 'age', bins=30)
g.add_legend()
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Age Distribution by Income Level')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='age', y='hours.per.week',hue='income',  alpha=0.6)
plt.title('Age vs Hours Worked per Week by Income Level')
plt.show()
```

# Summary of Findings

1. **Data Structure**:
   - The dataset contains categories like age, education, marital status, gender, hours worked per week, and income level (income >50K or <=50K).
   - There are a variety of education levels, marital statuses, and a mix of male and female respondents.

2. **Missing Values**:
   - There may be some missing values or inconsistent entries (e.g., represented as `?`). These need to be cleaned for accurate analysis.

3. **Demographic Distribution**:
   - The age distribution indicates a significant number of individuals in the 30-50 age range, with fewer individuals under 20 or over 70.
   - A considerable portion of the dataset consists of individuals with a high school education or less, but there are also significant numbers with college degrees.

4. **Income Distribution**:
   - The count plots for income reveal a significant imbalance, with more individuals earning less than 50K compared to those earning more than 50K.
   - Education plays a crucial role in income levels; individuals with higher education levels tend to have a higher income.

5. **Hours Worked**:
   - The box plots and histograms show that individuals working more than 40 hours per week are more likely to earn more than 50K.
   - There are clear outliers in the hours worked, particularly in the higher income bracket, indicating that higher working hours often correlate with higher income.

6. **Gender Analysis**:
   - There is a noticeable gender gap in income levels, with men more likely to earn above 50K compared to women. This reflects broader societal trends in income disparity.

7. **Marital Status**:
   - The relationship between marital status and income indicates that married individuals, especially those with spouses earning income, tend to have higher incomes.

8. **Correlation Insights**:
   - Numerical feature correlations highlight that age and hours worked are positively correlated with income, while education has a strong categorical impact on income levels.

# Insights

1. **Impact of Education**:
   - Higher education significantly increases the likelihood of earning above 50K. This suggests that educational attainment is a key factor in improving income prospects.

2. **Age and Work Experience**:
   - As individuals age, they generally have more work experience, which is likely reflected in their income. This indicates that age and accumulated experience play an essential role in income growth.

3. **Gender Disparities**:
   - The income gap between genders highlights the need for targeted policies and initiatives to address wage disparities in the workforce. Employers should focus on equitable pay practices.

4. **Working Hours**:
   - The correlation between higher working hours and income suggests that individuals may need to work longer hours to achieve higher income, raising questions about work-life balance and employee well-being.

5. **Marital Status Considerations**:
   - The data indicates that being married may have financial benefits, potentially due to dual incomes.
